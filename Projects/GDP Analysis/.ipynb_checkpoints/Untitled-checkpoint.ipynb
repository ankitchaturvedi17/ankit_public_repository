```python
filelist = os.listdir(os.getcwd()+'\\GSVA') 
```

```python
dfb = pd.read_csv(os.getcwd()+'\\GSVA\\'+filelist[0],encoding = "ISO-8859-1")
m = dfb.iloc[:,1].values
```

```python
df = pd.DataFrame(index=m)
```

```python
for x in filelist:    
    dfb = pd.read_csv(os.getcwd()+'\\GSVA\\'+x,encoding = "ISO-8859-1")   
    m = dfb.iloc[:,5].values
    df[x[4:-21]] = m   
```

```python
df.tail()
```

```python
df.sort_values(by=['Per Capita GSDP (Rs.)'],axis=1,inplace=True)
```

```python
plt.figure(figsize=(10, 8))
sns.barplot(x=df.iloc[32,:].values, y=df.columns.values)
plt.xlabel('GDP per Capita')
plt.ylabel('States')
plt.title('GDP per capita for all the states')
plt.show()
```

### Based on GDP per Capita
>Top 5 States Are Goa, Sikkim, Haryana, Kerela, Uttarakhand <br>
>Bottom 5 States are Bihar, Uttar Pradesh, Manipur, Assam , Jharkhand
### Ratio of Higest vs Lowest GDP is approximately 8:1
> Highest belong to Goa that is 271793
> and Lowest belong to Bihar 33954


```python
df2 = pd.DataFrame(index=df.columns.values)
df2['Primary'] = df.loc['Primary'].values
df2['Secondary'] = df.loc['Secondary'].values
df2['Tertiary'] = df.loc['Tertiary'].values
df2['TOTAL GSVA at basic prices'] = df.loc['TOTAL GSVA at basic prices'].values
```

```python
df2.index.name = 'States'
```

```python
df2.sort_values(by=['TOTAL GSVA at basic prices'],axis=0,inplace=True)
```

```python
df3= df2.iloc[:,0:3]
```

### Creating a new dataframe to calculate percentage of each category

```python
primarypercent = [(x/(x+y+z))*100 for x,y,z in zip(df3['Primary'].values,df3['Secondary'].values,df3['Tertiary'].values) ]
Secondarypercent = [(y/(x+y+z))*100 for x,y,z in zip(df3['Primary'].values,df3['Secondary'].values,df3['Tertiary'].values) ]
Tertiarypercent = [(z/(x+y+z))*100 for x,y,z in zip(df3['Primary'].values,df3['Secondary'].values,df3['Tertiary'].values) ]
```

```python
df8 = pd.DataFrame(index=df3.index.values)
df8['Primary'] = primarypercent
df8['Secondary'] = Secondarypercent
df8['Tertiary'] = Tertiarypercent
```

```python
df8.plot(stacked=True, width=1, kind="barh", lw=1, figsize=(10, 8))
```

### Above Plot represents percentage contribution of primary, secondary and tertiary sectors as a percentage of total GDP for all the states.

```python
quantiles = pd.qcut(df.iloc[32,:], q= [0,0.20,0.5, 0.85,1],labels=['C4','C3','C2','C1'])
```

### Below we are categorizing states based on Quantiles

```python
plt.figure(figsize=(10, 8))
sns.barplot(x=df.iloc[32,:].values, y=df.columns.values,hue=quantiles)
plt.xlabel('GDP per Capita')
plt.ylabel('States')
plt.title('GDP per capita for all the states')
plt.show()
```

### The above graph shows quantiles containing C1,C2,C3,C4 categories based on (0.20,0.5, 0.85, 1) quantiles where 85th and the 100th percentile are in C1, those between 50th and 85th percentile are in C2 and so on

```python
df.tail()
```

```python
dfread = pd.read_csv(os.getcwd() +'\\Dropout rates\\data.csv')
dfread.head()
```

### Clening and taking out the required data

```python
df2 = dfread.iloc[np.r_[1:5,6,10:14,15:18,19:26,27:35],np.r_[1, 3,7,10,13]]
```

```python
states = [word.replace(' ','_') for word in df2['Level of Education - State']]
```

```python
df2['Level of Education - State'] = states
```

```python
df6 = df.transpose()
df6 = df6.iloc[:,[32]]
```

```python
df2
```

### We will be working with df2 now

```python
corr = pd.merge(df2, df6, how="inner", left_on="Level of Education - State", right_on=df6.index)

corr.head()
```

### The above dataset will be used to find the correlaation between dropouts and GDP per capita 

```python
corr2= corr.iloc[:,1:]
corr2.head()
```

```python
cor = corr2.corr()
round(cor, 3)
```

### The above table represents the Correletion Table between Dropouts and GDP per capita

### Representing the above table in the form of Heatmap below

```python
plt.figure(figsize=(10,8))
# heatmap
sns.heatmap(cor, cmap="YlGnBu", annot=True)
plt.show()
```

### This Heatmap represents the Correlation between Dropout rates compared to GDP per capita

### Negative relation between Primary, Upper Primary and Secondry shows that GDP is inversely propotional to the dropout rates in the same
